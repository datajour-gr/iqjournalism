### Eπεξεργασία Greek Subjectivity Lexicon

```python
import pandas as pd
```

#### Εισαγωγή του επιμελημένου excel με βάρη, polarity, pos etc

#### Ορίζω το **path** του excel αρχείου ΚΑΙ το **πλήθος** των sheets

```python
# path και όνομα exel αρχείου
excel_path = 'xlsx/Subjectivity_Lex_GR_Edited.xlsx'
# πληθος sheets
sheets_nbr = 8
```

#### Διαβάζω τα sheets του excel και σώζω τα dfs τους στη λίστα sheetsL

```python
# names_proper = ['Strength', 'Length', 'Word', 'Part-of-speech', 'Stemmed', 'Polarity']
# names_in_excel = ['Type', 'Strength', 'Len', 'Word', 'Original', 'Pos', 'Part-of-speech', 'Stemmed', 'Polarity']
```

```python
sheetsL = []
for i in range(0, sheets_nbr):
    sheet_df = pd.read_excel(excel_path, sheet_name=i)
    sheetsL.append(sheet_df)
    print('sheet' + str(i+1) +':', len(sheet_df), 'rows')
```

```python
# Concatenate all the dataframes in the list
all_sheets_df = pd.concat(sheetsL, ignore_index=True)

all_sheets_df
```

#### Τσεκάρω για missing values

```python
for i in range(0, len(sheetsL)):
    try:
        print('sheet' + str(i+1), 'English:')
        print(sheetsL[i][sheetsL[i]['English'].isna()])
        print('-' * 10)
    except:
        pass
    try:
        print('sheet' + str(i+1), 'Greek:')
        print(sheetsL[i][sheetsL[i]['Greek'].isna()])
        print('-' * 10)
    except:
        pass
    
    print('=' * 20)
```

#### Τσεκάρω τα ονόματα των στηλών

```python
for i in range(0, len(sheetsL)):
    print('sheet' + str(i+1), 'Cols:')
    print(list(sheetsL[i][sheetsL[i].columns]))
```

```python
# κάνε string τυχόν αριθμούς
all_sheets_df['English'] = all_sheets_df['English'].astype('str')

# αφαίρεσε τα duplicate rows (τα rows που η αγγλική και η ελληνική λέξη είναι ίδιες)
new_sheet_df = all_sheets_df.drop_duplicates(subset=['English', 'Greek'],
                                                keep = 'first').reset_index(drop = True)
# ταξινόμησε το df ως προς τις αγγλικές λέξεις
all_sheets_df = all_sheets_df.sort_values('English').reset_index(drop=True)

all_sheets_df
```

#### Αποθήκευση του csv

```python
# αποθήκευσε το λεξικό σαν csv
all_sheets_df.to_csv('fixed_csvs/Greek_Subjectivity_Fixed_2023_02.csv', index=False)
```
