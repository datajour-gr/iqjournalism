###  Εμπλουτισμός λεξικου VAD με τις επιμελημένες μεταφράσεις από το Subjectivity Lexicon

Σε αυτό το Notebook συγχωνεύουμε το λεξικό VAD με τις μεταφρασεις που προέρχονται από το Subjectivity

```python
import pandas as pd
```

#### Ορίζω το **path** του excel αρχείου ΚΑΙ εισάγουμε το λεξικο VAD

```python
# path και όνομα exel αρχείου
excel_path = 'xlsx/Greek_NRC_VAD_Lex.xlsx'
# πληθος sheets
#sheets_nbr = 1
```

```python
df1 = pd.read_excel(excel_path)
```

```python
df1
```

#### Τσεκάρω για missing values

```python
missing_rows = df1[df1.isna().any(axis=1)]
missing_rows
#df1['English Word'].isna()
```

```python
df1 = df1.dropna()
```

```python
#μετονομάζουμε τις στήλες
df1 = df1.rename(columns={'English Word': 'English','Greek Word': 'Greek'})
```

```python
df1.head(30)
```

#### Εισάγουμε το αρχείο με τις μεταφρασμένες λέξεις

```python
df2 = pd.read_csv('fixed_csvs/fixed_translated_words_2023_02.csv')
df2.head(40)
```

```python
merged_df = pd.merge(df1, df2, on='English', how='left')
```

```python
merged_df.head(20)
```

```python
# Creating a list to store the final rows
final_rows = []

# Iterating through the merged dataframe
for index, row in merged_df.iterrows():
    if pd.isna(row['Valence']):  # If the weights are NaN, use the corresponding weights from df1
        match_row = df1[df1['English'] == row['English']].iloc[0]
        row['Valence'] = match_row['Valence']
        row['Arousal'] = match_row['Arousal']
        row['Dominance'] = match_row['Dominance']
    final_rows.append(row)

# Concatenating the rows into the final dataframe
final_df = pd.concat(final_rows, axis=1).T

# Resetting the index
final_df.reset_index(drop=True, inplace=True)

# Result
final_df.head(30)
```

```python
# Replace values in Greek_x with Greek_y where Greek_y is not NaN
final_df['Greek_x'] = final_df['Greek_y'].combine_first(final_df['Greek_x'])
```

```python
final_df.head(30)
```

```python
# Drop the Greek_y column
final_df.drop(columns=['Greek_y'], inplace=True)

# Rename the Greek_x column to 'Greek'
final_df.rename(columns={'Greek_x': 'Greek'}, inplace=True)
```

```python
final_df.head(30)
```

#### Αποθήκευση του csv

```python
# αποθήκευσε το λεξικό σαν csv
final_df.to_csv('fixed_csvs/edited_nrc_VAD.csv', index=False)
```
