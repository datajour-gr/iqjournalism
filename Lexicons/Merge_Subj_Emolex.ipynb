### Εμπλουτισμός λεξικου EMOLEX με τις επιμελημένες μεταφράσεις από το Subjectivity Lexicon

Σε αυτό το Notebook συγχωνεύουμε το  EMOLEX με τις μεταφρασεις που προέρχονται από το Subjectivity

```python
import pandas as pd
```

#### Ορίζω το **path** του excel αρχείου ΚΑΙ το **πλήθος** των sheets

```python
# path και όνομα exel αρχείου
excel_path = 'xlsx/Greek_NRC_EMOLEX-2.xlsx'
# πληθος sheets
#sheets_nbr = 1
```

#### Διαβάζω τα sheets του excel και σώζω τα dfs τους στη λίστα sheetsL

```python
df1 = pd.read_excel(excel_path)
```

```python
df1.head(30)
```

```python
df1.shape
```

#### Τσεκάρω για missing values

```python
missing_rows = df1[df1.isna().any(axis=1)]
missing_rows
```

```python
df1 = df1.dropna()
```

```python
df1 = df1.rename(columns={'English Word': 'English', 'Greek Word': 'Greek'})
```

```python
df1.head(30)
```

#### Φέρνω και το csv με τις μεταφρασμένες λέξεις

```python
df2 = pd.read_csv('fixed_csvs/fixed_translated_words_2023_02.csv')
df2.tail(40)
```

```python
merged_df = pd.merge(df1, df2, on='English', how='left')
```

```python
merged_df.head(30)
```

```python
# δημιουργία κενής λίστας
final_rows = []

# ταιριάζουμε ελληνικές λέξεις με τα συναισθηματικά σκορ
for index, row in merged_df.iterrows():
    if pd.isna(row['anger']):  # Αν τα βάρη είναι NaN, αφήνουμε τα βάρη του df1
        match_row = df1[df1['English'] == row['English']].iloc[0]
        row['anger'] = match_row['anger']
        row['anticipation'] = match_row['anticipation']
        row['disgust'] = match_row['disgust']
        row['fear'] = match_row['fear']
        row['joy'] = match_row['joy']
        row['negative'] = match_row['negative']
        row['positive'] = match_row['positive']
        row['sadness'] = match_row['sadness']
        row['surprise'] = match_row['surprise']
        row['trust'] = match_row['trust']
    final_rows.append(row)

# Concat τις σειρε΄ς
final_df = pd.concat(final_rows, axis=1).T

# Reset the index
final_df.reset_index(drop=True, inplace=True)
final_df.tail(30)
```

```python
# αντικαθιστουμε τις τιμές της στήλης Greek_x με αυ΄τές της στήλης Greek_y που είναι οι επιμελημένες λέξεις 
#(όχι τις τιμές NaN)
final_df['Greek_x'] = final_df['Greek_y'].combine_first(final_df['Greek_x'])
```

```python
final_df.tail(30)
```

```python
# Σβήουμε τη στήλη Greek_y
final_df.drop(columns=['Greek_y'], inplace=True)

# Rename τη στήλη Greek_x σε Greek'
final_df.rename(columns={'Greek_x': 'Greek'}, inplace=True)
```

#### Αποθήκευση του csv

```python
final_df.to_csv('fixed_csvs/edited_nrc_EMOLEX.csv', index=False)
```
